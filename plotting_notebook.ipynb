```python
path = 'C:/Users/jonze/Desktop/bomberman_ml/mle-bomberman/agent_code/fc_agent/score_per_round.txt'
with open(path, 'r') as f:
    data = f.read()

data = data.split('\n')
data = [line.split('\t') for line in data]
data = data[:-1]

data = [line[3:] for line in data] # ignore iter, round, epsilon

# get labels from first line
labels = data[0]

# get data from other lines
data = data[1:]

# create plots
import matplotlib.pyplot as plt
import numpy as np
# create bar plots for each label
# create a plot with 5 subplots that plot the data
fig, axs = plt.subplots(7, 1, figsize=(10, 10))
fig.tight_layout(pad=3.0)


# add the subplots to the plot
print(labels)
for i, label in enumerate(labels):
    axs[i].set_title(label)
    axs[i].set_xlabel('round')
    if label in ["avg_loss_per_step", "avg_reward_per_step", "invalid_actions_per_round", "avg_invalid_actions_per_step"]:
      float_data = [float(value) for value in np.array(data)[:, i]]
      axs[i].plot(float_data)
    else:
      axs[i].plot(np.array(data)[:, i])

plt.show()
```
